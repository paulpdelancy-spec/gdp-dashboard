```python
# 🛡️ SOVEREIGN ENGINE v44.1: MASTER INTEGRATED STEWARDSHIP CORE
# =================================================================
# MERGED: Essential 48 + Multimedia + Stripe Billing + 4hr Sync
# Optimized for: April 2026 | Yeshua AI Apps | Strategic Growth
# =================================================================

!pip install -q -U google-genai==1.64.0 google-auth==2.47.0 newspaper3k lxml_html_clean edge-tts fpdf2 mutagen pytz

import os, datetime, time, pytz, smtplib, asyncio, random
import edge_tts
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
from google.colab import userdata
from google.genai import Client
from newspaper import Article
from fpdf import FPDF
from mutagen.mp3 import MP3
from mutagen.id3 import ID3, APIC, TIT2, TPE1, error
import IPython.display as ipd

# =================================================================
# 🛡️ SECTION 1: STRIPE SUBSCRIPTION MATRIX
# =================================================================
STRIPE_MATRIX = """
--- 🛡️ STEWARDSHIP SUBSCRIPTION PLANS ---
1 Report/Day: $10/mo (https://buy.stripe.com/9B66oIcTW85N16o4xe8g00a) | $120/yr (https://buy.stripe.com/7sY5kE5rufyf9CUe7O8g00b)
2 Reports/Day: $19/mo (https://buy.stripe.com/fZu7sMcTW5XF9CUgfW8g00c) | $228/yr (https://buy.stripe.com/4gM8wQ5rubhZ4iAd3K8g00d)
3 Reports/Day: $28/mo (https://buy.stripe.com/aFabJ25rubhZ16obZG8g00e) | $336/yr (https://buy.stripe.com/8x25kE3jmcm3dTa4xe8g00f)
4 Reports/Day: $36/mo (https://buy.stripe.com/5kQ8wQ2fi0DleXe0gY8g00g) | $432/yr (https://buy.stripe.com/aFa4gA8DGadV8yQfbS8g00h)
5 Reports/Day: $45/mo (https://buy.stripe.com/4gM6oIdY04TB16o1l28g00i) | $540/yr (https://buy.stripe.com/6oUbJ2aLOcm37uM3ta8g00j)
6 Reports/Day: $54/mo (https://buy.stripe.com/6oU8wQ4nqfyfaGYe7O8g00k) | $648/yr (https://buy.stripe.com/3cIdRa8DG4TBeXe3ta8g00l)
"""

# =================================================================
# 🛡️ SECTION 2: DATA NODE REGISTRY
# =================================================================
FINANCIAL_NODES = ["www.cmegroup.com", "www.nyse.com", "www.nasdaq.com", "www.cboe.com", "www.bloomberg.com", "www.reuters.com/finance", "www.lseg.com/en", "www.hkex.com.hk", "www.jpx.co.jp/english", "www.sse.com.cn", "www.nseindia.com", "www.tmx.com", "www.euronext.com/en", "www.deutsche-boerse.com", "www.six-group.com/en", "www.sgx.com", "www.asx.com.au", "www.bseindia.com", "www.jse.co.za/trade", "www.idx.co.id/en-us", "www.set.or.th/en/home", "www.bursamalaysia.com", "www.twse.com.tw/en", "www.nadex.com"]
SPORTS_NODES = ["www.espn.com", "www.nfl.com", "www.nba.com", "www.premierleague.com", "www.skysports.com", "www.cbssports.com", "www.mlb.com", "www.nhl.com", "www.bundesliga.com", "www.laliga.com/en-GB", "www.legaseriea.it/it", "ligue1.com", "www.mlssoccer.com", "www.iplt20.com", "www.afl.com.au", "www.nrl.com", "www.ligaportugal.pt", "eredivisie.nl", "www.cfl.ca", "www.ligamx.net", "www.cbf.com.br", "www.jleague.jp/en", "sports.sina.com.cn/csl", "www.afa.com.ar"]

# =================================================================
# 🛡️ SECTION 3: THE MASTER AGENT (EZRA & JESSA)
# =================================================================
class StewardshipAgent:
    def __init__(self):
        self.api_key = userdata.get('GEMINI_API_KEY')
        self.client = Client(api_key=self.api_key)
        self.model_id = "gemini-2.0-flash"
        self.memory_log = "stewardship_memory.txt"
        self.agent_image = "image_0.png.jfif"

    def process_nodes(self, nodes, sector_name):
        full_analysis = ""
        chunks = [nodes[i:i + 8] for i in range(0, len(nodes), 8)]

        history = ""
        if os.path.exists(self.memory_log):
            with open(self.memory_log, "r") as f: history = f.read()[-3000:]

        for idx, chunk in enumerate(chunks):
            print(f"🚜 {sector_name}: Batch {idx+1}/{len(chunks)}...")
            batch_data = ""
            for url in chunk:
                try:
                    article = Article(f"https://{url}")
                    article.download(); article.parse()
                    batch_data += f"\n[{url}]: {article.text[:400]}"
                except: batch_data += f"\n[{url}]: Shielded"

            prompt = f"Role: Ezra & Jessa. Analyze: {batch_data}\nPoints: 2/site. Context: {history}"

            for attempt in range(5):
                try:
                    response = self.client.models.generate_content(model=self.model_id, contents=prompt)
                    full_analysis += response.text + "\n"
                    break
                except:
                    wait = random.randint(75, 125)
                    print(f"⏳ Sabbath Rest: {wait}s (Attempt {attempt+1}/5)")
                    time.sleep(wait)

            if idx < len(chunks) - 1: time.sleep(random.randint(15, 30))
        return full_analysis

    async def generate_assets(self, report_text):
        a_file, p_file = "briefing.mp3", "report.pdf"
        # Append Stripe Info to the Audio
        full_audio_text = report_text + "\nTo adjust your subscription frequency, please refer to the stewardship links in the attached PDF."
        await edge_tts.Communicate(full_audio_text[:8000], "en-US-AndrewNeural").save(a_file)

        # Branded PDF with Stripe Links
        pdf = FPDF(); pdf.add_page()
        if os.path.exists(self.agent_image):
            pdf.image(self.agent_image, x=10, y=10, w=190)
            pdf.ln(115)
        pdf.set_font("helvetica", "B", 16)
        pdf.cell(0, 10, "Sovereign Stewardship Briefing", new_x="LMARGIN", new_y="NEXT", align='C')
        pdf.set_font("helvetica", size=10); pdf.ln(5)
        pdf.multi_cell(0, 8, report_text.encode('latin-1', 'replace').decode('latin-1'))

        # New Page for Billing
        pdf.add_page()
        pdf.set_font("helvetica", "B", 14)
        pdf.cell(0, 10, "Subscription & Stewardship Management", new_x="LMARGIN", new_y="NEXT")
        pdf.set_font("helvetica", size=9)
        pdf.multi_cell(0, 6, STRIPE_MATRIX)
        pdf.output(p_file)

        return a_file, p_file

    def send_stewardship_email(self, a_file, p_file):
        sender = "paul.pdelancy@gmail.com"; pwd = userdata.get('GMAIL_APP_PASSWORD')
        recipients = [sender, "paulacumberbatch@yahoo.com"]
        msg = MIMEMultipart()
        msg['Subject'] = f"🛡️ Sovereign Pulse: {datetime.datetime.now().strftime('%m/%d %I:%M %p')}"
        msg['From'] = f"Ezra & Jessa <{sender}>"; msg['To'] = ", ".join(recipients)

        body = f"High-Reliability Stewardship Audit Complete.\n\n{STRIPE_MATRIX}\n\nSigned Management\nGod Bless"
        msg.attach(MIMEText(body, 'plain'))

        for f_path in [a_file, p_file]:
            with open(f_path, "rb") as f:
                part = MIMEApplication(f.read(), Name=os.path.basename(f_path))
                part['Content-Disposition'] = f'attachment; filename="{os.path.basename(f_path)}"'
                msg.attach(part)
        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
            server.login(sender, pwd); server.sendmail(sender, recipients, msg.as_string())
        print("📧 Briefing Dispatched with Integrated Billing.")

# =================================================================
# 🛡️ SECTION 4: THE ETERNAL SYNCHRONIZED LOOP
# =================================================================
async def run_master_engine():
    master_email = "paul.pdelancy@gmail.com"
    if userdata.get('AUTHORIZED_OWNER_EMAIL') != master_email:
        print("❌ Authorization Denied."); return

    agent = StewardshipAgent()
    est = pytz.timezone('US/Eastern')
    print("🚀 Sovereign Engine v44.1 Ignited. Starting Master Loop...")

    while True:
        # 1. RUN AUDIT
        finance_rep = agent.process_nodes(FINANCIAL_NODES, "FINANCE")
        sports_rep = agent.process_nodes(SPORTS_NODES, "SPORTS")
        full_report = f"GLOBAL STEWARDSHIP REPORT\n{finance_rep}\n{sports_rep}"

        # 2. GENERATE & SEND
        audio, pdf = await agent.generate_assets(full_report)
        agent.send_stewardship_email(audio, pdf)
        ipd.display(ipd.Audio(audio, autoplay=True))

        # 3. CALCULATE NEXT PULSE (4-HOUR SYNC)
        now = datetime.datetime.now(est)
        sched = [2, 6, 10, 14, 18, 22]
        next_h = min([h for h in sched if h > now.hour], default=2)
        next_p = now.replace(hour=next_h, minute=0, second=0, microsecond=0)
        if next_h == 2 and now.hour >= 22: next_p += datetime.timedelta(days=1)

        wait_seconds = (next_p - now).total_seconds()
        print(f"📊 Pulse Sent. Next Audit at: {next_p.strftime('%I:%M %p')} EST.")
        await asyncio.sleep(wait_seconds)

await run_master_engine()












```
